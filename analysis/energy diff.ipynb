```python
import networkx as nx
import numpy as np
import pickle
from parser import *
import os
import matplotlib.pyplot as plt
```

```python
MERAs_fn = "MERAs_64_disorder.p"
data_dir = "../data/last_few_bits_L_64_energy_avg/"
W_list = range(1,11)
num_files = 100
```

```python
def load_diagnostics(fname):
    E_dict = {}
    with open(fname, 'r') as fp:
        fp.readline()
        for i in range(5):
            data = np.array(fp.readline().split()).astype(np.float)
            E_dict[5 - i] = data[1:]
    return E_dict
```

```python
W_info = {}
dE_list = []
for W in W_list:
    i_list = []
    dE = []
    for i in range(num_files):
        d = "{}W_{}/".format(data_dir, W)
        fname = "{}W_{}_{:02}.txt".format(d, W, i)
        E_dict = load_diagnostics(fname)
        i_list.append(E_dict)
        dE.append(E_dict[1][1] - E_dict[1][0])
    dE_list.append((np.mean(dE), np.std(dE)))
    W_info[W] = i_list
```

```python
%matplotlib notebook
[means, stds] = list(zip(*dE_list))
plt.errorbar(W_list, means, yerr=stds)
plt.xlabel("W")
plt.ylabel(r"$\Delta E$", rotation=0)
plt.show()
```

```python
def get_gaps(N, n):
    gap_means = []
    gap_stds = []
    for W in W_list:
        E_W = W_info[W]
        i_list = []
        for i in range(num_files):
            E_dict = E_W[i]
            E_list = E_dict[N]
            E_diffs = np.abs(np.array(E_list[1:]) - np.array(E_list[:-1]))
            #print(E_diffs)
            gap_stats = [min(E_diffs[i:i+2])/max(E_diffs[i:i+2])for i in range(len(E_diffs)-2)]
            if gap_stats[n] <= 1 and gap_stats[n] >=0:
                i_list.append(gap_stats[n])
            #i_list += [g for g in gap_stats if g <= 1 and g >= 0]
        gap_means.append(np.mean(i_list))
        gap_stds.append(np.std(i_list)/np.sqrt(len(i_list)))
    return gap_means, gap_stds
        
```

```python
%matplotlib notebook
N = 5
for n in range(2**N-3):
    gap_means, gap_stds = get_gaps(N,n)
    plt.errorbar(W_list, gap_means, yerr=gap_stds, label="n={}".format(n+2))
plt.legend(loc='best')
plt.xlabel("W")
plt.ylabel(r"$\frac{E_{min}}{E_{max}}$", rotation=0)
plt.show()
```
