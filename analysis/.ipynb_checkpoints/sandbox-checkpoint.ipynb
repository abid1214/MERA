```python
L = 32
U_list = [None]*L
p_list = [None]*L
filename = "W_10.txt"
```

```python
with open(filename, 'r') as fp:
    fp.readline()
    i = 0
    for line in fp:
        data = list(map(int, line.split()[:-1]))
        U_list[i]=data
        i += 1
```

```python
U_list
```

```python
import networkx as nx
G = nx.Graph()
```

```python
for i in range(L):
    G.add_node(i)
    p_list[i] = i
```

```python
for k, U in enumerate(U_list):
    lbl = "U" + str(k)
    G.add_node(lbl)
    for i in range(U[0], U[2] + 1):
        G.add_edge(p_list[i], lbl)
        p_list[i] = lbl
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.figure()
nx.draw_spring(G, with_labels=True, font_weight='bold')
```

```python
sp = dict(nx.all_pairs_shortest_path(G))
```
